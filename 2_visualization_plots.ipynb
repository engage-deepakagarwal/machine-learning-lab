```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/hotel_bookings.csv')
```

# Line Plot
- When to use: To show trends over time or continuous data changes.
- Example: 
  - Imagine you are tracking how far in advance guests book their hotel rooms (that's the `lead_time`, measured in days). 
  - Now, instead of looking at individual dates, you group these bookings by week numbers of the year (`arrival_date_week_number`). 
  - This grouping allows you to see if people book earlier during certain weeks (e.g., summer vacation weeks or holiday seasons) or closer to the date during less busy times.

```python
plt.plot(
    df['arrival_date_week_number'], 
    df['lead_time'], 
    marker='o', 
    color='purple'
)
plt.xlabel('Week Number')
plt.ylabel('Lead Time (in days)')
plt.show()
```

Data points are highly dense, making it difficult to distinguish individual values and trends.

```python
# Aggregate data: Calculate the mean lead time for each week number
aggregated_data = df.groupby('arrival_date_week_number')['lead_time'].mean().reset_index()
plt.plot(
    aggregated_data['arrival_date_week_number'], 
    aggregated_data['lead_time'], 
    marker='o', 
    color='purple'
)
plt.xlabel('Week Number')
plt.ylabel('Lead Time (in days)')
plt.show()
```

# Histogram
- Scenario: Understanding the distribution of numerical variables.
- Example: Distribution of lead_time.

Understanding:
- Think of the histogram as sorting people into "buckets" based on how many days ahead they booked.
- Each bucket (bar) represents a range of days (e.g., 050 days).
- The height of the bar shows how many bookings are in that bucket.

For example:
- If the bar for "050 days" is the tallest, it means most people booked within that time frame.
- If theres a short bar for "600700 days," it means very few people booked that far in advance.

```python
import matplotlib.pyplot as plt

# Histogram for lead_time
plt.hist(df['lead_time'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Lead Time', fontsize=14)
plt.xlabel('Lead Time (days)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

# Bar chart
- Scenario: Comparing categorical data frequencies.
- Example: Frequency of different hotel types.

Understanding in the context
- It shows how many hotels are registered per category

```python
# Bar plot for hotel types
df['hotel'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title('Frequency of Hotel Types', fontsize=14)
plt.xlabel('Hotel Type', fontsize=12)
plt.ylabel('Count of hotels', fontsize=12)
plt.show()
```

# Pie Chart
- Scenario: Proportions of reservation_status.
- Example: Showing the percentage split of reservation statuses.

```python
# Pie chart for reservation status
df['reservation_status'].value_counts().plot(
    kind='pie'
)
plt.title('Reservation Status Proportions', fontsize=14)
plt.show()
```

# Box Plot
- Scenario: Identifying outliers and the spread of numerical data.
- Example: adr (Average Daily Rate) comparison by hotel type.

```python
import seaborn as sns

# Box plot for ADR by hotel type
sns.boxplot(data=df, x='hotel', y='adr')
plt.title('ADR Comparison by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('ADR (Average Daily Rate)')
plt.show()
```

# Scatter Plot
- Scenario: Correlation between two numerical features.
- Example: ADR (Average Daily Rate) vs. lead_time.

```python
# Scatter plot for ADR vs Lead Time
plt.scatter(df['lead_time'], df['adr'])
plt.title('ADR vs Lead Time')
plt.xlabel('Lead Time (days)')
plt.ylabel('ADR (Average Daily Rate)')
plt.show()
```

# Heatmap
- Scenario: Visualizing correlations between numerical features.
- Example: Correlation matrix of numerical columns.

```python
# Heatmap for correlation matrix
import seaborn as sns

sns.heatmap(df.select_dtypes(include='number').corr())
plt.title('Correlation Matrix')
plt.show()
```
